```python

# ============================================================================
# SECTION 3: FEATURE ENGINEERING
# ============================================================================


print("FEATURE ENGINEERING")


# Create a copy for feature engineering
df_fe = df.copy()

# 1. Time-based features
df_fe['Order_Year'] = df_fe['Order Date'].dt.year
df_fe['Order_Month'] = df_fe['Order Date'].dt.month
df_fe['Order_Quarter'] = df_fe['Order Date'].dt.quarter
df_fe['Order_DayOfWeek'] = df_fe['Order Date'].dt.dayofweek
df_fe['Order_DayOfYear'] = df_fe['Order Date'].dt.dayofyear

# 2. Shipping lag (days between order and ship)
df_fe['Shipping_Days'] = (df_fe['Ship Date'] - df_fe['Order Date']).dt.days

# 3. Price-related features
df_fe['Price_Per_Unit'] = df_fe['Sales'] / df_fe['Quantity']
df_fe['Profit_Margin'] = df_fe['Profit'] / df_fe['Sales']
df_fe['Discount_Amount'] = df_fe['Sales'] * df_fe['Discount']

# 4. Sales per unit after discount
df_fe['Sales_Per_Unit_After_Discount'] = df_fe['Price_Per_Unit'] * (1 - df_fe['Discount'])

# 5. Profitability indicators
df_fe['Is_Profitable'] = (df_fe['Profit'] > 0).astype(int)
df_fe['Profit_Category'] = pd.cut(df_fe['Profit'], 
                                   bins=[-np.inf, 0, 50, 100, np.inf],
                                   labels=['Loss', 'Low', 'Medium', 'High'])

# 6. Sales volume indicator
df_fe['Is_High_Volume'] = (df_fe['Quantity'] >= df_fe['Quantity'].quantile(0.75)).astype(int)

# 7. Aggregated customer features (RFM-like)
customer_stats = df_fe.groupby('Customer ID').agg({
    'Order ID': 'count',  # Frequency
    'Sales': ['sum', 'mean', 'std'],
    'Profit': ['sum', 'mean'],
    'Quantity': 'mean'
}).reset_index()

customer_stats.columns = ['Customer ID', 'Purchase_Frequency', 'Total_Customer_Sales', 
                          'Avg_Order_Value', 'Std_Order_Value', 'Total_Customer_Profit', 
                          'Avg_Customer_Profit', 'Avg_Quantity_Per_Order']

df_fe = df_fe.merge(customer_stats, on='Customer ID', how='left')

# 8. Categorical encoding (for now, keep as is for reference)
df_fe['Segment_Code'] = df_fe['Segment'].astype('category').cat.codes
df_fe['Region_Code'] = df_fe['Region'].astype('category').cat.codes
df_fe['Category_Code'] = df_fe['Category'].astype('category').cat.codes

print("\nNew features created:")
print(df_fe.columns.tolist())

print("\n" + "=" * 50)
print("FEATURE STATISTICS")
print("=" * 50)
display(df_fe[[col for col in df_fe.columns if col not in df.columns]].describe())

print("\nMissing values in engineered features:")
print(df_fe[[col for col in df_fe.columns if col not in df.columns]].isnull().sum())
```
