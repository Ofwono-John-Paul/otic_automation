```python
from dotenv import load_dotenv
import os

load_dotenv() 
```

```python
# Read the cleaned Otic text from file
with open("otic.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Step 2: Chunk the text
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = text_splitter.split_text(raw_text)

print(f"Total chunks: {len(chunks)}")
print("Sample chunk:\n", chunks[0])
```

```python
import shutil
import os

# Delete HF cache (optional safety step)
shutil.rmtree(os.path.expanduser("~/.cache/huggingface/hub"), ignore_errors=True)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document

# Step 3: Prepare documents
docs = [Document(page_content=chunk) for chunk in chunks]

# Step 4: Load a reliable model (works without auth and downloads cleanly)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Build and save vector store
db = FAISS.from_documents(docs, embeddings)
db.save_local("otic_vectorstore")
```

```python
db = FAISS.load_local(
    "otic_vectorstore",
    embeddings,
    allow_dangerous_deserialization=True  # ✅ you trust this source
)
retriever = db.as_retriever()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA

# Initialize Gemini LLM from env key
llm = ChatGoogleGenerativeAI(
    model="models/gemini-1.5-flash-latest",  # or "gemini-pro-1.5"
    temperature=0.1
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
```

```python
query = "Who is Mr. Paul Katende in OTIC Foundation?"
result = qa_chain.run(query)
print("Answer:", result)
```

```python
print("Welcome to the OTIC Interaction Bot!")
print("Type your question below. Type 'quit' or 'exit' to end.\n")

while True:
    query = input("You > ")

    if query.strip().lower() in ["quit", "exit"]:
        print("👋 Goodbye! Thanks for interacting with the OTIC Bot.")
        break

    try:
        result = qa_chain.run(query)
        print("OTIC Bot >", result)
    except Exception as e:
        print("⚠️ Sorry, something went wrong:", str(e))
-
```
